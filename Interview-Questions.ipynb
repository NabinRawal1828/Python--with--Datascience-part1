```python
you are working on a project to analyze weather data. create variables to store the current temperature, and the minimum temperature 
recorded. Determine whether the current temperature is within the range of the minimum and maximum temperatures recorded using logical opeartors.
```

```python

current_temp = float(input("Enter the current temperature: "))
maximum_temp = float(input("Enter the maximum temperature: "))
minimum_temp = float(input("Enter the minimum temperature: "))

if (current_temp >= minimum_temp) and (current_temp <= maximum_temp):
    print("The temperature is within range")
else:
    print("The temperature is not within range")
```
