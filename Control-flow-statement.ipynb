# if statement

```python
If statement is a control statement used to check a condition and execute code only if the condition is true.
```

```python
n = 20
if(n > 10):
    print("20 is greater than 10")
else:
    print("20 is less than 10")
```

```python
Develop a temperature converter in python for a cold storage that converts temperatures between Celsius and Fahrenheit scales.
The program should take a temperature value as input (in Celsius).
If the temperature is less than 15°C, inform the user that the temperature is not convenient (too cold).
If the temperature is between 24 to 28°C, inform the user that the temperature is convenient (optimal range).
```

```python
temp = float(input("Enter the temperature:"))
if temp < 15:
    print("The temperature is not convenient (Too cold)")
elif 24 <= temp <= 28:
    print("Temperature is convenient (Optimal range)")
else:
    print("The temperature is above the optimal range")
```


# if else statement

```python
 If–else statement executes one block if the condition is true, otherwise it executes another block.
```

```python
Write a program to check whether a person is eligible for voting or not
```

```python
age = int(input("Enter your age for voting: "))
if age >= 18:
    print("You are eligible for voting")
else:
    print("You are not eligible for voting")
```

```python
You are tasked with developing a python program to manage employee salaries for a company. your program should calculate 
the net salarly of each employee based on their base salarly, deductions, and bonuses.Additionally,employees who have been with
the company for more than 5 years, 8% of salarly.Deduction of tax will be 12%

write a python script that prompts the user to input the following information for each employee:
Basic salarly
Years of service

After calculating the net salarly, the program should print a summary for each employee including their base salarly,
deductions, bounses, loyality bonus (if applicable), and net salarly
```

```python
base_salary = float(input("Enter your base salary: "))
years_of_service = int(input("Enter your years of service: "))

# Calculate loyalty bonus
loyalty_bonus = 0.08 * base_salary if years_of_service > 5 else 0

# Calculate tax deduction
tax_deduction = 0.12 * base_salary

# Calculate net salary
net_salary = base_salary + loyalty_bonus - tax_deduction

print("\n--- Salary Summary ---")
print(f"Base Salary: Rs {base_salary}")
print(f"Loyalty Bonus (8% if >5 years): Rs {loyalty_bonus}")
print(f"Tax Deduction (12%): Rs {tax_deduction}")
print(f"Net Salary: Rs {net_salary}")
```

```python
write a python program that prompt the user to input a ipl teams and displays their famous players.
```

```python
ipl_team = input("Enter the IPL team name (e.g., Chennai, RCB, Mumbai, SRH): ").lower()

if ipl_team == "chennai":
    print("Famous Player: MS Dhoni")
elif ipl_team == "rcb":
    print("Famous Player: Virat Kohli")
elif ipl_team == "mumbai":
    print("Famous Player: Rohit Sharma")
elif ipl_team == "srh":
    print("Famous Player: Abishek Sharma")
else:
    print("Invalid team! Please enter a valid IPL team name.")
```
