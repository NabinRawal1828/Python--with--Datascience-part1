# Question - 3

```python
You're developing a financial application that calculate simple interest.
The principal amount is 250000 Rs, the interest rate is 9.34567%
and the time period is 3 years.Calculate the simple interest (SI),
and the final value should be integer.
    
```

```python
p = 250000
R = 9.34567
T = 3
SI = (p*T*R)/100
print("The simple interest is :",SI)
print("The Interger value is:",int(SI))
```

# Question - 3

```python
Write a python code for hostel students, ask them to mention their course name and age. for 5 students and calculate their age in avg.
```

```python
ages = []
courses = []
for i in range(1,6):
    age = int(input("Enter the age of students:"))
    ages.append(age)
    course = input("Enter your course name:")
    courses.append(course)
    
print("The ages of 5 students is :",ages)
print("The courses of the students is :",courses)

length = len(ages)
total = sum(ages)
averages = total/length
print("averages of 5 students ",averages)


```
